```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
indexed_time = pd.read_csv("indexed.csv")
coord_time = pd.read_csv("coord.csv")
```

```python
for size, group in coord_time.groupby("Size"):
		plt.plot(group["Density"], group["Time (s)"], label=f"Coord {size}")
for size, group in indexed_time.groupby("Size"):
		plt.plot(group["Density"], group["Time (s)"], label=f"Indexed {size}")
plt.xlabel("Density")
plt.ylabel("Time (s)")
plt.legend()
```
